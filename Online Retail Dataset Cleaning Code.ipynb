# Cleaning the Online Retail Data Set

```python
import pandas as pd
```

```python
# Loading the dataset into a DataFrame
df = pd.read_excel('/home/chemweno/Desktop/TATA Data Project/Online Retail Data Set.xlsx')
df.head
```

```python
df.sample(15)
```

# Data Exploration
## Number of rows,columns and data types

```python

num_rows, num_cols = df.shape
data_types = df.dtypes

print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
print("Data types:")
print(data_types)

```

## Changing the data type for customerID

```python
# Changing the CustomerID column to string
df['CustomerID'] = df['CustomerID'].astype(str).str.replace('\.0', '')
df.dtypes
```

## Descriptive statistics of the numerical data types

```python
# Finding the descriptive statistics for the Quantity and UnitPrice columns
df[['Quantity', 'UnitPrice']].describe()
```

```python
df.sample(15)
```

# Checking for missing data 


```python
# Checking for missing values
missing_data = df.isnull().sum()
print(missing_data)
```

```python
# Filter the DataFrame to show rows where Description is missing
missing_description = df[df['Description'].isnull()]

# Display the rows where Description is missing
print("Rows where Description is missing:")
print(missing_description)
```

```python
# Filter the DataFrame to show rows where Description is missing
missing_description.sample(15)
```

# Removing the missing values
## Started using df_cleaned here

```python
df_cleaned = df.dropna()
df_cleaned.describe()
```

```python
# Check for missing values after removal
missing_data_after_removal = df_cleaned.isnull().sum()

# Display the number of missing values after removal
print("Missing data after removal:")
print(missing_data_after_removal)
```

# Checking for duplicate rows

```python
duplicate_rows = df_cleaned.duplicated()

# Filter the DataFrame to show duplicate rows
duplicate_rows_df = df_cleaned[duplicate_rows]

# Display the duplicate rows
print("Duplicate rows:")
print(duplicate_rows_df)
```

```python
# Count the number of duplicate rows
duplicate_rows_count = duplicate_rows.sum()

# Display the number of duplicate rows
print("Number of duplicate rows:", duplicate_rows_count)
```

## Excluding rows from the DataFrame df_cleaned where the quantity is negative and where the unit price is below $0.

```python
# Exclude rows with negative quantities
df_cleaned = df_cleaned[df_cleaned['Quantity'] >= 1]

# Exclude rows with unit prices below $0
df_cleaned = df_cleaned[df_cleaned['UnitPrice'] >= 0]
```

```python
df_cleaned.describe()
```
